```python
# read tsv
import pandas as pd
from pathlib import Path
import os

BASE_DIR = Path.cwd()
CLEANED_DIR = BASE_DIR / '..' / 'cleaned' / 'flores.csv'

cleaned_df = pd.read_csv(CLEANED_DIR, sep=',')
complete_df = pd.read_csv(BASE_DIR / '..' / 'flores_plus' / 'parallel.tsv', sep='\t')

display(cleaned_df.head())
display(complete_df.head())
```

```python
print(cleaned_df.info())
print(complete_df.info())
```

```python
# Extract id from cleaned_df
cleaned_df['id'] = cleaned_df['original_id'].apply(lambda x: int(x.split('_')[1]))

# Identify the new columns we want to bring over from complete_df
new_columns = ['sentence_id', 'croato', 'occitano', 'portoghese', 'ungherese']

# Merge the dataframes
merged_df = pd.merge(
    cleaned_df, 
    complete_df[new_columns], 
    left_on='id', 
    right_on='sentence_id', 
    how='left'
).drop(columns=['id', 'sentence_id'])

# Update cleaned_df
cleaned_df = merged_df

display(cleaned_df.head())
```

```python
# overwrite csv
cleaned_df.to_csv(CLEANED_DIR, index=False)
```
